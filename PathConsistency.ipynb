```python
miningFile = open("data/mining.csv", 'r')
miningData = miningFile.readlines()
miningFile.close()

defaultFile = open("data/standard.csv", 'r')
defaultData = defaultFile.readlines()
defaultFile.close()

miningData = [float(i.strip()) for i in miningData]
defaultData = [float(i.strip()) for i in defaultData]
```

```python
def getDeltas(data):
    deltas = []
    for i in range(1, len(data)):
        deltas.append(data[i] - data[i-1])
    return deltas
```

```python
miningDeltas = getDeltas(miningData)
defaultDeltas = getDeltas(defaultData)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(18, 10), dpi=96)
plt.plot(defaultDeltas, color='royalblue')
plt.ylabel('Button Press Delta')
plt.ylim([6.2390, 6.2420])
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()

figure(figsize=(18, 10), dpi=96)
plt.plot(miningDeltas, color='chocolate')
plt.ylabel('Button Press Delta')
plt.ylim([6.2390, 6.2420])
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
import numpy as np
print("STANDARD")
print("Mean: " + str(np.mean(defaultDeltas)))
print("Std: " + str(np.std(defaultDeltas)))
print("Worst From Mean: " + str(max(max(defaultDeltas) - np.mean(defaultDeltas), np.mean(defaultDeltas) - min(defaultDeltas))))
print("Max - Min: " + str(max(defaultDeltas) - min(defaultDeltas)))

print("MINING")
print("Mean: " + str(np.mean(miningDeltas)))
print("Std: " + str(np.std(miningDeltas)))
print("Worst From Mean: " + str(max(max(miningDeltas) - np.mean(miningDeltas), np.mean(miningDeltas) - min(miningDeltas))))
print("Max - Min: " + str(max(miningDeltas) - min(miningDeltas)))
```
